# Brief introduction 

This notebook provides the structure factors $S_{\rm p}$, $S_{\rm n}$, $S_{\rm 00}$, $S_{\rm 01}$ and $S_{\rm 11}$ for the spin-dependent WIMP-nucleus scattering in $^{19}$F, $^{23}$Na, $^{27}$Al, $^{29}$Si, $^{73}$Ge, $^{127}$I, and $^{129,131}$Xe as published in Ref. [1]. In addition, we provide ab-initio results of spin expectation for proton $\mathbf{S}_{\rm p}$ and neutron $\mathbf{S}_{\rm n}$. 

---
The results are obtained from the following ab-initio calculations: starting from a set of established two- (NN) and three-nucleon (3N) Hamiltonians, derived within chiral effective field theory (EFT), we construct corresponding WIMP-nucleon currents at the one-body level and include the effects from axial-vector two-body currents. We then apply the ab initio in-medium similarity renormalization group (IMSRG) to construct effective valence-space Hamiltonians and consistently transformed operators of nuclear responses.

* The differential cross section for spin-dependent WIMP elastic scattering off a nucleus in the ground state with total angular momentum $J$ is given by
\begin{equation}
\frac{d \sigma}{d q^{2}}=\frac{8 G_{\mathrm{F}}^{2}}{(2 J+1) v^{2}} S_{A}(\mathbf{q}^{2}),
\end{equation}
where $\mathbf{q}$ ($q \equiv|\mathbf{q}|$) denotes the momentum transfer from nucleus to WIMP, $v$ indicates the WIMP velocity, and $S_{A}(\mathbf{q}^2)$ is the axial-vector structure factor, obtained from detailed nuclear theory calculations.

* The structure factor $S_{A}(\mathbf{q})$ can be decomposed into isoscalar/isovector components $S_{00}$, $S_{01}$ and $S_{11}$ through
\begin{equation}
S_{A}(\mathbf{q})=a_{0}^{2} S_{00}(\mathbf{q})+a_{0} a_{1} S_{01}(\mathbf{q})+a_{1}^{2} S_{11}(\mathbf{q}),
\end{equation}
where the labels 0($+$) and 1($-$) distinguish the isoscalar and isovector contributions, respectively, and the coupling constants $a_0$ and $a_1$ are form factors accounting for information from the particle and hadronic sectors.
It is common in the literature to use the proton-only $S_{\rm p}$ and neutron-only $S_{\rm n}$ which are defined by the couplings $a_0=a_1=1$ and $a_0=-a_1=1$, respectively.

* At $q$=0 MeV, the structure factor is given by:
\begin{equation}
\begin{array}{l}
S_{A}(0)=\dfrac{(J+1)(2J+1)}{4 \pi J}\left| \left(a_{0}+a_{1}^{\prime}\right)\left\langle\mathbf{S}_{\rm p}\right\rangle+\left(a_{0}-a_{1}^{\prime}\right)\left\langle\mathbf{S}_{\rm n}\right\rangle \right|^{2}
\end{array}
\end{equation}
where $\mathbf{S}_{\rm p}$ and $\mathbf{S}_{\rm n}$ denote the total spin expectation for protons and neutrons, respectively, and $a_{1}^{\prime}=a_1(1+ \delta a(0))$ includes the effects from chiral two-body currents (2BCs), namely $\delta a(0)$.

# Structure factors $S_{\rm p}$, $S_{\rm n}$, $S_{\rm 00}$, $S_{\rm 01}$ and $S_{\rm 11}$ 

This notebook outputs $S_{\rm p}$, $S_{\rm n}$, $S_{\rm 00}$, $S_{\rm 01}$ and $S_{\rm 11}$ with three different NN+3N interactions from chiral EFT with consistent 2BCs: 'DNNLOgo', 'EM1.8_2.0', 'N4LO_LNL2'. 

You need to <span style='background:yellow'> set three input parameters </span>: 
<span style='color:Red'> nuclei </span>, 
<span style='color:Red'> interactions </span>, 
<span style='color:Red'> qs </span>.

The nuclei can be:
'F19', 'Na23', 'Al27', 'Si29', 'Ge73', 'I127', 'Xe129', 'Xe131'

namely,

<span style='color:Red'> nuclei = ['F19']  </span>
or 
<span style='color:Red'> nuclei = ['F19', 'Xe131']  </span>


The interactions can be:
'DNNLOgo', 'EM1.8_2.0', 'N4LO_LNL2'

namely,

<span style='color:Red'> interactions=['DNNLOgo']  </span>
or
<span style='color:Red'> interactions=['DNNLOgo', 'EM1.8_2.0', 'N4LO_LNL2']  </span>

The momentum transfer q in MeV can be:

<span style='color:Red'> qs = [0.0, 10.0, 20.0, 30.0]  </span> 
or 
<span style='color:Red'> qs = np.linspace(0,349,350)  </span>.

Note: the maximum q is 400 MeV.


The 2BC contributions are obtained via  normal-ordering with respect to Fermi gas reference state with density $\rho$ (see Ref.[10] for detailed expressions). 

We take the density range $\rho$=0.09...0.12 fm$^{-3}$. The notebook can change the density range by setting rhos, such as rhos = [0.1] or rhos = [0.09, 0.095, 0.100, 0.105, 0.110, 0.115, 0.120]. 

[10] M. Hoferichter, J. Menéndez, and A. Schwenk, [Phys. Rev. D 102, 074018 (2020)](https://journals.aps.org/prd/abstract/10.1103/PhysRevD.102.074018)


```python
import sys
print(sys.path[0])
from math import *
import cmath

# import the physical and mathematical constants
import DM_constants as dmc
# import the functions for 2BCs 
import DM_func as dmf

# Standard Python imports
import os, re
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
```

```python
### input nuclei: 'F19', 'Na23', 'Al27', 'Si29', 'Ge73', 'I127', 'Xe129', 'Xe131'
# nuclei = ['Xe129', 'Xe131']
nuclei = ['F19', 'Na23', 'Al27', 'Si29', 'Ge73', 'I127', 'Xe129', 'Xe131']

### input interactions: 'DNNLOgo', 'EM1.8_2.0', 'N4LO_LNL2'
# interactions=['EM1.8_2.0']
interactions=['DNNLOgo', 'EM1.8_2.0', 'N4LO_LNL2']

### input momentum transfer q in MeV. The maximum q is 400 MeV.
### qs = [0.0, 10.0, 20.0, 30.0] or qs = np.linspace(0,349,350)
qs = np.linspace(0,349,100)

### initialize rhos = [0.09, 0.095, 0.100, 0.105, 0.110, 0.115, 0.120] 
rhos = [0.09,0.120]
# rhos = [0.09, 0.095, 0.100, 0.105, 0.110, 0.115, 0.120]
```

```python

path_fit = sys.path[0] + '/data_fit'

Dic_vs = {
'F19':'sd-shell', 'Na23':'sd-shell', 'Al27':'sd-shell', 'Si29':'sd-shell',
'Ge73':'pf5g9-Ni56', 'I127':'sdg7h11-Sn100','Xe129':'sdg7h11-Sn100', 'Xe131':'sdg7h11-Sn100'
}

Dic_IMSRG = {
'F19': 'magnus_F19_e10_E22_hw16_A19',
'Na23': 'magnus_Na23_e10_E22_hw16_A23',
'Al27': 'magnus_Al27_e10_E22_hw16_A27',
'Si29': 'magnus_Si29_e10_E22_hw16_A29',
'Ge73': 'magnus_Ge73_e10_E22_hw16_A73',
'I127': 'magnus_I127_e14_E22_hw16_A127_eimsrg8_e2imsrg16_NAT',
'Xe129': 'magnus_Xe129_e14_E22_hw16_A129_eimsrg8_e2imsrg16_NAT',
'Xe131': 'magnus_Xe131_e14_E22_hw16_A131_eimsrg8_e2imsrg16_NAT',
}

df_SFs = pd.DataFrame(columns=['nuclei','interaction','q','2BCs','rho', \
'Sp','Sn','S00','S01','S11'])


for target in nuclei:
    IMSRG = Dic_IMSRG[target]
    hw_extract = re.findall(r'%s(\d+)'%'hw', IMSRG)
    hw = int(hw_extract[0])
    b_HO = np.sqrt(dmc.HBARC**2/dmc.M_NUCLEON/hw)
    for inte in interactions:
        if inte == 'EM1.8_2.0' : c1=-0.81; c3=-3.2; c4=5.4; cD=1.264
        elif inte == 'N4LO_LNL2' : c1=-1.1; c3=-5.54; c4=4.17; cD=-1.8
        elif inte == 'DNNLOgo' : c1=-0.74; c3=-3.622246; c4=2.446123; cD=0.081
        else: print("No default LECs for interaction " + ARGS['LECs']); break;
        c6 = 3.7040
 
        F_dSigmap = []; F_dSigmapp = [];
        name = path_fit+'/'+target+'_'+Dic_vs[target]+'_'+inte+'_'+IMSRG
        fit_sigmap = np.genfromtxt(fname=name+'_sigmap.dat')
        fit_sigmapp = np.genfromtxt(fname=name+'_sigmapp.dat')
        fit_n = 6
        if(len(fit_sigmap.shape) == 1):
            fit_sigmap = np.column_stack((fit_sigmap,0.0*fit_sigmap))
            fit_sigmapp = np.column_stack((fit_sigmapp,0.0*fit_sigmapp))
        for q in qs:
            u = q*q*b_HO*b_HO/2.0/dmc.HBARC/dmc.HBARC
            for irho, rho in enumerate(rhos):
                delta_a1, delta_a1_P = dmf.delta_2BC(float(rho),float(c1),float(c3),float(c4),float(c6),float(cD),q)
                dSigmap = dmf.dSigmap_1B(q) + delta_a1
                dSigmapp = dmf.dSigmapp_1B(q) + delta_a1 + delta_a1_P
                Sp_1B_fit = 0.0; Sn_1B_fit = 0.0;
                S00_1B_fit = 0.0; S01_1B_fit = 0.0; S11_1B_fit = 0.0;
                Sp_fit = 0.0; Sn_fit = 0.0;
                S00_fit = 0.0; S01_fit = 0.0; S11_fit = 0.0;

                for i_fit in range(0, fit_sigmap.shape[1]):
                    FSigmap = []; FSigmapp = []
                    for j_fit in range(0,2):
                        ci = fit_sigmap[j_fit*fit_n:(j_fit+1)*fit_n, i_fit]
                        FSigmap.append( dmf.Fit_F(u, ci) )
                        ci = fit_sigmapp[j_fit*fit_n:(j_fit+1)*fit_n, i_fit]
                        FSigmapp.append( dmf.Fit_F(u, ci) )

                    Fi = [FSigmap[0], FSigmap[1], FSigmapp[0], FSigmapp[1]]
                    Sp_fit +=dmf.F2S('p', Fi, dSigmap, dSigmapp)
                    Sn_fit +=dmf.F2S('n', Fi, dSigmap, dSigmapp)
                    S00_fit +=dmf.F2S('00', Fi, dSigmap, dSigmapp)
                    S01_fit +=dmf.F2S('01', Fi, dSigmap, dSigmapp)
                    S11_fit +=dmf.F2S('11', Fi, dSigmap, dSigmapp)
                    Sp_1B_fit += dmf.F2S('p', Fi, dmf.dSigmap_1B(q), dmf.dSigmapp_1B(q))
                    Sn_1B_fit += dmf.F2S('n', Fi, dmf.dSigmap_1B(q), dmf.dSigmapp_1B(q))
                    S00_1B_fit += dmf.F2S('00', Fi, dmf.dSigmap_1B(q), dmf.dSigmapp_1B(q))
                    S01_1B_fit += dmf.F2S('01', Fi, dmf.dSigmap_1B(q), dmf.dSigmapp_1B(q))
                    S11_1B_fit += dmf.F2S('11', Fi, dmf.dSigmap_1B(q), dmf.dSigmapp_1B(q))
                        
                if(irho == 0):
                    data_add = {'nuclei':target, 'interaction':inte, 'q':q, '2BCs':'without', 'rho':0.0, \
                    'Sp':Sp_1B_fit, 'Sn':Sn_1B_fit, 'S00':S00_1B_fit, 'S01':S01_1B_fit, 'S11':S11_1B_fit}
                    df_SFs = df_SFs.append(data_add,ignore_index=True)
                data_add = {'nuclei':target, 'interaction':inte, 'q':q, '2BCs':'with', 'rho':rho, \
                'Sp':Sp_fit, 'Sn':Sn_fit, 'S00':S00_fit, 'S01':S01_fit, 'S11':S11_fit}
                df_SFs = df_SFs.append(data_add,ignore_index=True)
            
                    
print(df_SFs.columns)                           
                
```

```python
### set wo2BC = False or True
### If wo2BC = False, the plots will show both results with 2BCs and without 2BCs. 
### While wo2BC = True, the plots will only show results with 2BCs.
wo2BC = True
# wo2BC = False


for itarget, target in enumerate(nuclei):
    for inte in interactions:
        df_temp = df_SFs[(df_SFs['nuclei'] == target) & (df_SFs['interaction'] == inte)]
        if (wo2BC):
            wo2BC_index = df_temp [ df_temp ['2BCs'] == 'without' ].index
            df_temp = df_temp .drop(wo2BC_index)
        plt.semilogy(df_temp['q'],df_temp['Sp'],'o',ms=6)
        plt.semilogy(df_temp['q'],df_temp['Sn'],'>',ms=6)
    if(target[2:3].isdigit()): AA = target[2:]
    if(target[1:2].isdigit()): AA = target[1:]
    if(target[2:3].isdigit()): nucl_AA = target[0:2]
    if(target[1:2].isdigit()): nucl_AA = target[0]
    plt.text(0.5,0.8,'$^{%s}$%s'%(AA,nucl_AA),transform=plt.gca().transAxes,fontsize=16)
    plt.xlabel('$q$ (MeV)',fontsize=14)
    plt.ylabel('$S_{\\rm{i}}$($q^2$)',fontsize=14)
    plt.show()
    
plt.show()
```
